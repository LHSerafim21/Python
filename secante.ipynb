```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def secante(f, a, b, tol = 1e-4, maxIt=50):
    if f(a)*f(b) > 0:
        print("O intervalo dado é inválido")
        return None, 0
    
    if np.abs(f(a)) < tol:
        return a,0
    if np.abs(f(b)) < tol:
        return b,0
    
    for it in range(maxIt):
        c = (a+b)/2

        if np.abs(f(c)) < tol:
            return c, it+1

        if f(c)*f(a) < 0:
            b = c # intervalo [a,c]
        else:
            a = c # intervalo [c,b]
        
        c = (a*f(b) - b*f(a))/(f(b)-f(a))
    print('Numero max de iterações foi atingido')
    return c, maxIt
```

```python
def f(x):
    return np.exp(x) - x - 3*x

# intervalo de visualização
a = -2
b = 2

#plota intervalo par visualização
pts = np.linspace(a, b, 100)
plt.plot(pts, f(pts))
plt.hlines(0,a,b,'k')
```

```python
# Execução do algorítimo
x, it = secante(f, a, b, 1e-5)
print(f'A solução aproximada encontrada foi x = {x};\nValor f(x) = {f(x)};\nForam necessárias {it} iterações')
```
