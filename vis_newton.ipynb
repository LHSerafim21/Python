```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
from IPython import display
```

```python

# Define a função e sua derivada

def f(x):
    return x**3 - x**2 + x - 1

def df(x):
    return 3*x**2 - 2*x + 1

# Define o método de Newton
def newton(x0):
    x = x0
    fx = f(x)
    dfx = df(x)
    x_new = x - fx / dfx
    return x_new

# Define o intervalo de x e plota a função
x = np.linspace(-2, 2, 100)
y = f(x)
fig, ax = plt.subplots()
ax.plot(x, y)
x0 = -0.5
plt.plot(x0, f(x0), 'ro')
plt.hlines(0,-2, 2,'k')

# Define a linha tangente à função em um ponto x
def tangent_line(x0):
    y0 = f(x0)
    slope = df(x0)
    intercept = y0 - slope*x0
    return slope*x + intercept

# Define a função de animação
def animate(i):
    # Escolhe um valor inicial para x e executa o método de Newton
    x_newton = [x0]
    for j in range(i):
        # Calcula a linha tangente à função em x
        ax.clear()
        ax.plot(x, y)
        ax.hlines(0,-2, 2,'k')
        y_tan = tangent_line(x_newton[-1])
        
        # Plota a linha tangente
        ax.plot(x, y_tan, '--')

        # Plota as linhas tangentes anteriores
        for xi in x_newton[:-1]:
            y_tan = tangent_line(xi)
            ax.plot(x, y_tan, '--k', alpha = 0.3)
        
        # Calcula a nova raiz usando o método de Newton
        x_new = newton(x_newton[-1])
        x_newton.append(x_new)
        
        # Plota a nova raiz
        ax.plot(x_new, 0, 'ro')
        ax.plot(x_newton[-2], f(x_newton[-2]), 'go')
    
    # Define os limites dos eixos para destacar a área de interesse
    ax.set_xlim([-2, 2])
    ax.set_ylim([-3, 3])
    
    # Define o título do gráfico
    ax.set_title(f"Iteração {i+1}")
    
    # Remove as linhas tangentes das iterações anteriores
    # ax.lines = ax.lines[:1]
    
    return ax.lines

# Cria a animação
anim = animation.FuncAnimation(fig, animate, frames=10, interval=1500, blit=True)

# Mostra a animação
# plt.show()
plt.close()
video = anim.to_html5_video()
html = display.HTML(video)
display.display(html)
```

```python
# Define a função e sua derivada
def f(x):
    return x**3 - x**2 - 1

def df(x):
    return 3*x**2 - 2*x

# Define o intervalo de visualização
a = -7 # x_min
b = 7 # x_max
c = -10 # y_min
d = 10 # y_max

# Define o método de Newton
def newton(x0):
    x = x0
    fx = f(x)
    dfx = df(x)
    x_new = x - fx / dfx
    return x_new

# Define o intervalo de x e plota a função
x = np.linspace(a, b, 100)
y = f(x)
fat = (d-c)/(b-a)
fig, ax = plt.subplots()
ax.plot(x, y)
x0 = -0.5
plt.plot(x0, f(x0), 'ro')
plt.hlines(0,a, b,'k')

# Define a linha tangente à função em um ponto x
def tangent_line(x0):
    y0 = f(x0)
    slope = df(x0)
    intercept = y0 - slope*x0
    return slope*x + intercept

# Define a função de animação
def animate(i):
    # Escolhe um valor inicial para x e executa o método de Newton
    x_newton = [x0]
    for j in range(i): # executa i iterações de newton
        # Calcula a linha tangente à função em x
        ax.clear()
        ax.plot(x, y)
        ax.hlines(0,a, b,'k')
        y_tan = tangent_line(x_newton[-1])
        
        # Plota a linha tangente
        ax.plot(x, y_tan, '--')

        # Plota as linhas tangentes anteriores
        for xi in x_newton[:-1]:
            y_tan = tangent_line(xi)
            ax.plot(x, y_tan, '--k', alpha = 0.3)
        
        # Calcula a nova raiz usando o método de Newton
        x_new = newton(x_newton[-1])
        x_newton.append(x_new)
        # Plota a nova raiz
        ax.plot(x_new, 0, 'ro',label='$x_k$')
        ax.plot(x_newton[-2], f(x_newton[-2]), 'go')
    
    # Define os limites dos eixos para destacar a área de interesse
    ax.set_xlim([a, b])
    ax.set_ylim([c, d])
    
    # Define o título do gráfico
    ax.set_title(f"Iteração {i+1}")
    ax.legend()
    plt.tight_layout()
    # Remove as linhas tangentes das iterações anteriores
    # ax.lines = ax.lines[:1]
    
    return ax.lines

# Cria a animação
anim = animation.FuncAnimation(fig, animate, frames=30, interval=1000, blit=True)

# Mostra a animação
# plt.show()
plt.close()
video = anim.to_html5_video()
html = display.HTML(video)
display.display(html)
```
