```python
import numpy as np # funções matemáticas
import matplotlib.pyplot as plt # plot dos gráficos (visualização)

def bissec(f, a, b, tol = 1e-8, maxIt = 50):

    # Verificar condição de existencia de raiz no intervalo
    if f(a)*f(b) > 0:
        print("O intervalo dado é inválido")
        return None, 0 # retorna nenhuma raiz
    
        # alternativamente, podemos lançar uma exceção
        # raise ValueError(f"O intervalo [{a}, {b}] é inválido para essa função")

    # Verifica se os extremos do intervalo são raízes
    if np.abs(f(a)) < tol:
        return a,0
    if np.abs(f(b)) < tol:
        return b,0
    
    for it in range(maxIt): # for(int it = 0; it < maxIt; it++)

        x = (a+b)/2 # Realiza a iteração (calcula o ponto médio)

        # Verifica se resultado é próximo suficiente da raiz
        if np.abs(f(x)) < tol:
            return x, it+1
        
        # Determina novo intervalo
        if f(x)*f(a) < 0:
            b = x # intervalo [a,x]
        else:
            a = x # intervalo [x,b]

    # Caso não haja convergência no limite de iterações definido:
    print('Numero max de iterações foi atingido')
    return x, maxIt

def f(x):
    return x**2 - x - 1

# Intervalo de x
x = np.linspace(-2, 2, 100)

# Valores correspondentes de f(x)
y = f(x)

# Plot do gráfico da função
plt.plot(x, y)
plt.axhline(y=0, color='black', linestyle='--')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico de f(x) = x^2 - x - 1')
plt.grid(True)
plt.show()

print(f'Para achar os valores de busca é necessário plotar o gráfico e analisar aonde a curva da f(x) corta o eixo x. Após isso, precisa-se definir intervalos aonde dentro deles está esses pontos. Os pontos em questão podem ser analisados no gráfico abaixo, e os intervalos estabelecidos são: [-1, 0] e [1, 2]\n')
# Execução do algorítimo
x, it = bissec(f, -1, 0)
print(f'A solução aproximada encontrada foi x = {x};\nValor f(x) = {f(x)};\nForam necessárias {it} iterações')
# Execução do algorítimo
x, it = bissec(f, 1, 2)
print(f'A solução aproximada encontrada foi x = {x};\nValor f(x) = {f(x)};\nForam necessárias {it} iterações')

print(f'O critério de parada utilizado é verificar a proximidade entre f(x) e 0. Lembrando de utilizar a tolerância, que no caso desse exercício é 8 casas decimais')
```

```python
def interpolação_lagrange(x, y, xi):
    n = len(x)
    resultado = 0.0

    for i in range(n):
        termo = y[i]
        for j in range(n):
            if j != i:
                termo *= (xi - x[j]) / (x[i] - x[j])
        resultado += termo

    return resultado

# Exemplo de uso
x = [1.4, 1.5, 1.6, 1.7, 1.8, 1.9, 2.0]                         # Valores conhecidos de x
y = [-3.3375, -4.199, -5.07, -5.989, -6.958, -7.979, -9.054]    # Valores correspondentes de y

xi = 1.85          # Ponto em que queremos interpolar

valor_interp = interpolação_lagrange(x, y, xi)
print("Valor interpolado em xi =", valor_interp)

#----------------------------------------------------------------------------------------------------------------------
import math

def calcular_erro_limitante(x, y, xi, n):
    # Cálculo da derivada de ordem (n+1) da função f(x)
    def derivada(x):
        return -4 * math.sin(x) - math.exp(x)

    # Cálculo do fatorial de (n + 1)
    fatorial = math.factorial(n + 1)

    # Cálculo do produto dos termos (xi - xj)
    produto = 1
    for j in range(len(x)):
        if x[j] != xi:
            produto *= (xi - x[j])

    # Cálculo do limitante do erro
    erro_limitante = abs(derivada(xi) / fatorial * produto)

    return round(erro_limitante, 3)


# Exemplo de uso
x = [1.4, 1.5, 1.6, 1.7, 1.8, 1.9, 2.0]  # Valores conhecidos de x
y = [-3.3375, -4.199, -5.07, -5.989, -6.958, -7.979, -9.054]  # Valores correspondentes de y

xi = 1.85  # Ponto em que queremos interpolar
grau = 3   # Grau do polinômio interpolador

erro_limitante = calcular_erro_limitante(x, y, xi, grau)
print("Limitante do erro de P3(1.85):", erro_limitante)

```
